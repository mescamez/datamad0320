```python
df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname AS "FIRST NAME", 
t.title AS "TITLE", pub.pub_name AS "PUBLISHER"
FROM authors AS au INNER JOIN titleauthor AS tta ON au.au_id = tta.au_id
INNER JOIN titles as t ON tta.title_id = t.title_id
INNER JOIN publishers as pub ON t.pub_id = pub.pub_id;
""", engine)
```
