# Challenge1

```python
df = pd.read_sql_query("""
SELECT
    au.au_id AS "AUTHOR ID", 
    au.au_lname AS "LAST NAME", 
    au.au_fname AS "FIRST NAME", 
    t.title AS "TITLE", 
    pub.pub_name AS "PUBLISHER"
FROM authors AS au 
    INNER JOIN titleauthor AS tta ON au.au_id = tta.au_id
    INNER JOIN titles as t ON tta.title_id = t.title_id
    INNER JOIN publishers as pub ON t.pub_id = pub.pub_id;
""", engine)
```

# Challenge2

```python
df1 = pd.read_sql_query(""" 
SELECT
	    au.au_id AS "AUTHOR ID", 
        au.au_lname AS "LAST NAME", 
        au.au_fname AS "FIRST NAME", 
        pub.pub_name AS "PUBLISHER", 
        count(*) AS "TITLE COUNT"
FROM authors AS au
	INNER JOIN titleauthor AS tta ON au.au_id = tta.au_id
	INNER JOIN titles AS t ON tta.title_id = t.title_id
	INNER JOIN publishers AS pub ON t.pub_id = pub.pub_id
GROUP BY au.au_id, pub.pub_name;
""", engine)
```

```python
df1
```

# Challenge3

```python
df2 = pd.read_sql_query(""" 
SELECT
		au.au_id AS "AUTHOR ID", 
        au.au_lname AS "LAST NAME", 
        au.au_fname AS "FIRST NAME", 
        SUM(sl.qty) AS "TOTAL"
FROM authors AS au
	INNER JOIN titleauthor AS tta ON au.au_id = tta.au_id
	INNER JOIN titles AS t ON tta.title_id = t.title_id
    LEFT JOIN sales AS sl ON t.title_id = sl.title_id
GROUP BY au.au_id, au.au_lname, au.au_fname
ORDER BY TOTAL DESC
LIMIT 3;
""", engine)
```

```python
df2
```
